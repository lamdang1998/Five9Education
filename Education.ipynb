```python
import operator
def word_count(doc):
    dic = {}
    available = []
    for sentence in doc:
        sentence = remove_accents2(sentence)
        sentence = sentence.lower()
        for word in sentence.split():
            if word not in available:
                available.append(word)
                dic[word] = 1
            else:
                dic[word] = dic[word] + 1
    return dic

a1_sorted_keys = sorted(word_count(documents), key=word_count(documents).get, reverse=True)
for r in range (14):
    print (a1_sorted_keys[r], word_count(documents)[a1_sorted_keys[r]])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score


documents = load_data_db(db_edu)
documents = documents[:1000]


vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

true_k = 2
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1, random_state = 1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print

print("\n")
print("Prediction")

Y = vectorizer.transform(["truong thpt chuyen sp"])
prediction = model.predict(Y)
print(prediction)

Y = vectorizer.transform(["Trai tổ lái gái ham đua ôm cua ốp vỉa bốc đầu mài răng đá lửa"])
prediction = model.predict(Y)
print(prediction)
 
```
